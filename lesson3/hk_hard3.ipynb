```python
# Задание-1:
# Написать программу, выполняющую операции (сложение и вычитание) с простыми дробями.
# Дроби вводятся и выводятся в формате:
# n x/y ,где n - целая часть, x - числитель, у - знаменатель.
# Дроби могут быть отрицательные и не иметь целой части, или иметь только целую часть.
# Примеры:
# Ввод: 5/6 + 4/7 (всё выражение вводится целиком в виде строки)
# Вывод: 1 17/42  (результат обязательно упростить и выделить целую часть)
# Ввод: -2/3 - -2
# Вывод: 1 1/3


from fractions import Fraction

str = '5/6 + 4/7'
str = '-2/3 - -2' 
ls = str.split(sep=' ')
oper = ''
nd = 0
for indx,i in enumerate(ls):
    if i == '-' or i == '+':
        oper = i
        nd = indx
ch1 = ''
dh1 = ''
num1_z = '+'
num2_z = '+'
for i in range(0,nd):
    if ls[i].rfind('/') < 0:
        ch1 = ls[i]
    else:
        dh1 = ls[i]
# first fraction
if len(ch1) > 0:
    if ch1[0] == '-':
        num1_z = '-'
        ch1 = ch1.replace('-','')
    ch1 = Fraction(int(ch1),1)
else:
    ch1 = Fraction(0,1)
if len(dh1) > 0:
    if dh1[0] == '-':
        num1_z = '-'
        dh1 = dh1.replace('-','')
    dh1 = Fraction(int(dh1[0]),int(dh1[2]))
else:
    dh1 = Fraction(0,1)
# second fraction
ch2 = ''
dh2 = ''
for i in range(nd + 1, len(ls)):
    if ls[i].rfind('/') < 0:
        ch2 = ls[i]
    else:
        dh2 = ls[i]
if len(ch2) > 0:
    if ch2[0] == '-':
        num2_z = '-'
        ch2 = ch2.replace('-','')
    ch2 = Fraction(int(ch2),1)
else:
    ch2 = Fraction(0,1)
    
if len(dh2) > 0:
    if dh2[0] == '-':
        num2_z = '-'
        dh2 = dh2.replace('-','')
    dh2 = Fraction(int(dh2[0]),int(dh2[2]))
else:
    dh2 = Fraction(0,1)
num1 = ch1 + dh1
num2 = ch2 + dh2
if num1_z == '-':
    num1 *= -1
if num2_z == '-':
    num2 *= -1
if oper == '+':
    result = num1 + num2
if oper == '-':
    result = num1 - num2
print(result.numerator // result.denominator,result.numerator % result.denominator,'/',result.denominator)
    


    

```

```python
# Задание-2:
# Дана ведомость расчета заработной платы (файл "data/workers").
# Рассчитайте зарплату всех работников, зная что они получат полный оклад,
# если отработают норму часов. Если же они отработали меньше нормы,
# то их ЗП уменьшается пропорционально, а за заждый час переработки
# они получают удвоенную ЗП, пропорциональную норме.
# Кол-во часов, которые были отработаны, указаны в файле "data/hours_of"
fw = open('workers','r')
fh = open('hours_of','r')
ls = []
#for id,line in enumerate(fw):
#    if id > 0:
#        line = line.replace('\n','')
#        ls_work1.append(line.split(sep='\t'))
#ls_work1
for id,line in enumerate(fw):
    if id > 0:
        line = line.replace('\n','')
        ls.append(line.split(sep=' '))
ls_work = []        
for i in ls:
    l = []
    for j in i:
        if len(j) > 0:
            l.append(j)
    ls_work.append(l)
    
ls = []
for id,line in enumerate(fh):
    if id > 0:
        line = line.replace('\n','')
        ls.append(line.split(sep=' '))

ls_h = []        
for i in ls:
    l = []
    for j in i:
        if len(j) > 0:
            l.append(j)
    ls_h.append(l)

fw.close()
fh.close()

for id,i in enumerate(ls_work):
    n_w = ls_work[id][0]
    f_w = ls_work[id][1]
    s_w = ls_work[id][2]
    h_w = ls_work[id][4]
    for j,val in enumerate(ls_h):
        if ls_h[j][0] == n_w and ls_h[j][1] == f_w:
            if h_w == ls_h[j][2]:
                print('Сотрудник ',n_w,f_w,' получает зарплату ', s_w)
            elif ls_h[j][2] < h_w:
                salary_of_hour = round(float(s_w) / float(h_w))
                hours = int(h_w) - int(ls_h[j][2])
                print('Сотрудник ',n_w,f_w,' получает зарплату ', float(s_w) - float(hours)*salary_of_hour,'р')
            elif ls_h[j][2] > h_w:
                salary_of_hour = round(float(s_w) / float(h_w))
                hours = int(ls_h[j][2]) - int(h_w) 
                print('Сотрудник ',n_w,f_w,' получает зарплату ', float(s_w) + 2*float(hours)*salary_of_hour,'р')
```

```python
# Задание-3:
# Дан файл ("data/fruits") со списком фруктов.
# Записать в новые файлы все фрукты, начинающиеся с определенной буквы.
# Т.е. в одном файле будут все фрукты на букву “А”, во втором на “Б” и т.д.
# Файлы назвать соответственно.
# Пример имен файлов: fruits_А, fruits_Б, fruits_В ….
# Важно! Обратите внимание, что нет фруктов, начинающихся с некоторых букв.
# Напишите универсальный код, который будет работать с любым списком фруктов
# и распределять по файлам в зависимости от первых букв, имеющихся в списке фруктов.
# Подсказка:
# Чтобы получить список больших букв русского алфавита:
# print(list(map(chr, range(ord('А'), ord('Я')+1))))

f = open('fruit.txt','r')
lst_char = list(map(chr, range(ord('А'), ord('Я')+1)))
lst_fruit = []
for line in f:
    if len(line) > 1:
        line = line.replace('\n','')
        lst_fruit.append(line)
f.close()
for i in lst_char:
    lst_f_fruit = []
    for j in lst_fruit:
        if i == j[0]:
            lst_f_fruit.append(j)
    if len(lst_f_fruit) > 0:
        name = 'fruit_' + i
        fn = open(name,'w')
        for index in lst_f_fruit:
            fn.write(index + '\n')
        fn.close()
```
